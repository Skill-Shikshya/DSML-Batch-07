## Condition Based Filtering

````python
mask = titanic_df["Fare"] > 100

fare_100 = titanic_df[mask]
````

````python
fare_100.shape
````

````python
fare_100.head()
````

````python
female_df = fare_100[fare_100.Sex == "female"]
````

````python
female_df.head()
## Why nearly 65% of highest paying customer are female in 1912 ?
````

````python
survive = titanic_df[
    (titanic_df['Sex'] == "female") & (titanic_df['Survived'] == 1)
]
survive.shape
````

````python
survive_male = titanic_df[
    (titanic_df['Sex'] == "male") & (titanic_df['Survived'] == 1)
]
survive_male.shape
````

````python
survived_1 = titanic_df[
    (titanic_df["Pclass"] == 3) & (titanic_df["Survived"] == 1)
    ]
survived_1.shape
````
