**Handling Incorrect Data**

```python
import numpy as np

data = pd.DataFrame({
    "id":[1,3,5,7,9],
    "name":["John","Jane","Jennifer","Johan","Jeff"],
    "age":[25,-32,19,42,"55"],
    "gender" : ["M","M","Z","F","F"],
    "income":[400, 700, 300, np.nan, 600]
})
```

```python
data.dtypes
```

```python
data
```

```python
data.age = data.age.astype(int)
```

```python
# Sample Boundary
filter_age = data[(data.age > 18) & (data.age <= 45)]
filter_age
```

```python
data.age = abs(data.age)
data
```
