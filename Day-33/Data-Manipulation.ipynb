### Handling Null value for Age, using initials (Mr., Miss or Master)groups.

```python
titles = titanic_df['Name'].str.split(',')
titanic_df['Title'] = titles.str[1].str.strip().str.split(' ').str[0]
titanic_df.head()
```

```python
def get_titles(name):
    first_name = name.split(",")[-1]
    title = first_name.split(".")[0]
    return title.strip(" ")
```

```python
get_titles("Braund, Mr. Owen Harris")
```

```python
titanic_df['title'] = titanic_df.Name.apply(get_titles)
```

```python
titanic_df['title'].replace(
    to_replace=["Mlle","Ms","Lady"], value="Miss"
).value_counts()
```

```python
titanic_df.title.replace(
    to_replace={"Dr":"Mr","Col":"Mr"}
).value_counts()
```

```python
means_age = titanic_df.groupby("title")['Age'].transform('mean')
means_age
```

```python
titanic_df.Age = titanic_df['Age'].fillna(means_age)
```

```python
titanic_df.Age.plot.hist()
```
